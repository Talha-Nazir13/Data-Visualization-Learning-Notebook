# Customization of boxplot:

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df= sns.load_dataset("titanic")
df.head()
```

- simple plot

```python
sns.boxplot(x="class",
            y="fare",
           data=df)
plt.show()
```

- Show mean (green point shows mean):


```python
sns.boxplot(x="sex",
            y="age",
           data=df,
           showmeans=True,
           )
plt.show()
```

- add * instid of green point

```python
sns.boxplot(x="sex",
            y="age",
           data=df,
           showmeans=True,
            color="red",
            meanprops={"marker":"^","markersize":"15","markeredgecolor":"black"}
           )
plt.show()
```

```python
sns.boxplot(x="sex",
            y="age",
           data=df,
           showmeans=True,
            color="red",
            meanprops={"marker":"*","markersize":"15","markeredgecolor":"black"}
           )
plt.show()
```

- add Lebels on x axis and y axis of your choice
- change the fount size


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df= sns.load_dataset("titanic")
sns.boxplot(x="sex",
            y="age",
           data=df,
           showmeans=True,
            color="red",
            meanprops={"marker":"^","markersize":"15","markeredgecolor":"black"}
           )
# Add x axis and y axis
plt.xlabel(" Gender",size=20)
plt.ylabel("Age of Peopers",size=14)
# add title
plt.title("Boxplot of titanic base on age and sex",size=20)

plt.show()
```

- bold x and y Labels
- Add title and bold it

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df= sns.load_dataset("titanic")
sns.boxplot(x="sex",
            y="age",
           data=df,
           showmeans=True,
            color="red",
            meanprops={"marker":"^","markersize":"15","markeredgecolor":"black"}
           )
# Add x axis and y axis
plt.xlabel(" Gender",size=20,weight="bold")
plt.ylabel("Age of Peopers",size=20,weight="bold")
# add title
plt.title("Titanic data plot base on age and sex",size=20,weight="bold")

plt.show()
```

###  Relplot :

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# import dataset from seaborn
df=sns.load_dataset("dots")
# defining a color palette  
p=sns.color_palette("rocket_r")
# plot a lineplot
sns.relplot(
    data=df,
    x="time",
    y="firing_rate",
    hue="coherence",
    col="align",
    kind="line",
    size_order=["T1","T2"],
    palette=p,
    height=5,
    aspect=0.75,
    
    
)
plt.xlabel("TIME",size=10,weight="bold")
plt.ylabel("firing",size=10,weight="bold")
plt.show()
```

### Boxenplot:

```python
import seaborn as sns
sns.set_theme(style="dark",color_codes=True)
df=sns.load_dataset("tips")
print(df.head())
sns.boxenplot(data=df,
             x="day",
             y="total_bill",
             hue="smoker",
             palette={"No":"red","Yes":"blue"},
             )
plt.xlabel("DAY",size=20,weight="bold")
plt.ylabel("TOTAL_BILL",size=20,weight="bold")
plt.title("BOXENPlot Using tips dataset",size=20,weight="bold")
plt.show()
```

### Boxenplot:

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="dark",color_codes=False)
df=sns.load_dataset("dots")
sns.boxenplot(x="choice",y="coherence",
           color="b",
            scale='linear',
            data=df,
            palette="rocket_r"
        )
```

### violinplot:

```python
import seaborn as sns
import matplotlib.pyplot as plt
# set theme
sns.set_theme(style="white",color_codes=True)
# load dataset from seaborn
df=sns.load_dataset("tips")
print(df.head())
plt.title("violinplot using tips dataset",size=20,weight="bold")
sns.violinplot(data=df,
              x="day",
              y="total_bill",
              hue="smoker",
              split=True,
              linewidth=2,
              )
plt.show()

```

### scatterplot:

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks",color_codes=True)
# import Diamomonds dataset
df=sns.load_dataset("diamonds")
print(df.head())
plt.title("scatterplot using diamonds dataset",size=20,weight="bold")
sns.scatterplot(data=df,
              x="depth",
              y="table",
              linewidth=2,
              )
plt.show()
```

## Plotly :

### Plot 01:

```python
 import plotly.express as px
df=px.data.iris()
# print(df.head())
 df1=px.scatter(df,x="sepal_width",
                y="sepal_length",
                color="species",
                marginal_y="violin",
                marginal_x="box",
                trendline="ols",
                template="simple_white",
               
               )
df1.show()
```

### Plot 02:
- animated plot

```python
import plotly.express as px
df=px.data.gapminder()
# print(df.head())
df1=px.scatter(df,x="gdpPercap",
              y="lifeExp",
              animation_frame="year",
              animation_group="country",
              size="pop",
              color="continent",
              hover_name="country",
              facet_col="continent",
             log_x=True,
               size_max=45,
               range_x=[100,100000],
               range_y=[25,90]
               )
df1.show()
```

### Plot 03:

```python
import plotly.express as px
df=px.data.gapminder().query("year==2007").query("continent=='Europe'")
df.loc[df['pop']<2.e6,'country']='other countries'
df1=px.pie(df,values='pop',names='country',title='Population of European continent')
df1.show()
```

### Plot 04:

```python
import plotly.express as px
df=px.data.gapminder().query("year==2007")
df1=px.sunburst(df,path=["continent",'country'],values='pop',
               color='lifeExp',
               hover_data=["iso_alpha"])
df1.show()
```
